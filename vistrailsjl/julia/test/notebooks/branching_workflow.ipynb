# Branching Workflow Test

This workflow demonstrates branching where one module feeds multiple downstream modules.

## Workflow Structure

```
                    source ("Hello World")
                         |
           +-------------+-------------+
           |             |             |
        length        upper         reverse
           |             |             |
           |             +------+------+
           |                    |
           |                 combine
           |                    |
           +--------------------+
```

Expected results:
- length: 11
- upper: "HELLO WORLD"
- reverse: "dlroW olleH"
- combined: "HELLO WORLD + dlroW olleH"
- final: "11: HELLO WORLD + dlroW olleH" (combining length with combined string)

This requires converting the integer length to a string first.

````julia
#| module-id: length_to_string
#| module-type: julia:JuliaSource
#| inputs:
#|   length: length.length

length_val = get_input("length")
set_output("result", string(length_val))
````
