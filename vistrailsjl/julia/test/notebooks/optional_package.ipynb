# Optional Inputs Package

This package demonstrates modules with optional input ports.

Modules:
- **NumberSource**: Generates a number
- **OptionalAdd**: Adds two numbers, second number is optional (defaults to 0)
- **OptionalMultiply**: Multiplies by factor, factor is optional (defaults to 1)
- **FormatNumber**: Formats a number with optional prefix/suffix

## Package Metadata

```julia
#| package-meta
#| identifier: org.vistrails.vistrails.optional
#| version: 1.0.0
#| name: Optional Inputs Package
#| description: Modules demonstrating optional input ports
```

## NumberSource Module

Simple number generator.

```julia
#| module: NumberSource
#| output_ports:
#|   - name: value
#|     signature: basic:Float
#| parameters:
#|   - name: number
#|     signature: basic:Float

number = get_parameter("number")
set_output("value", number)
```

## OptionalAdd Module

Adds two numbers. The second number is optional and defaults to 0.

```julia
#| module: OptionalAdd
#| input_ports:
#|   - name: a
#|     signature: basic:Float
#|     label: first number
#|   - name: b
#|     signature: basic:Float
#|     label: second number
#|     optional: true
#| output_ports:
#|   - name: sum
#|     signature: basic:Float

a = get_input("a")
b = has_input("b") ? get_input("b") : 0.0
set_output("sum", a + b)
```

## OptionalMultiply Module

Multiplies a number by a factor. The factor is optional and defaults to 1.0.

```julia
#| module: OptionalMultiply
#| input_ports:
#|   - name: value
#|     signature: basic:Float
#|     label: input value
#|   - name: factor
#|     signature: basic:Float
#|     label: multiplication factor
#|     optional: true
#| output_ports:
#|   - name: result
#|     signature: basic:Float

value = get_input("value")
factor = has_input("factor") ? get_input("factor") : 1.0
set_output("result", value * factor)
```

## FormatNumber Module

Formats a number as a string with optional prefix and suffix.

```julia
#| module: FormatNumber
#| input_ports:
#|   - name: number
#|     signature: basic:Float
#|     label: number to format
#|   - name: prefix
#|     signature: basic:String
#|     label: prefix string
#|     optional: true
#|   - name: suffix
#|     signature: basic:String
#|     label: suffix string
#|     optional: true
#| output_ports:
#|   - name: formatted
#|     signature: basic:String

number = get_input("number")
prefix = has_input("prefix") ? get_input("prefix") : ""
suffix = has_input("suffix") ? get_input("suffix") : ""
set_output("formatted", prefix * string(number) * suffix)
```
