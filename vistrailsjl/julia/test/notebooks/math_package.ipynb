# Math Package

Simple arithmetic modules demonstrating Python VisTrails compatibility.

This package implements:
- **Divide**: Divides two numbers (with error handling)
- **Add**: Adds two numbers
- **Multiply**: Multiplies two numbers

These modules match the examples from Python VisTrails documentation.

```julia
#| package-meta
#| identifier: org.vistrails.vistrails.math
#| version: 1.0.0
#| description: Basic arithmetic operations
```

## Divide Module

Divides two floating-point numbers.

**Inputs**:
- `arg1` (Float): dividend
- `arg2` (Float): divisor

**Outputs**:
- `result` (Float): quotient

**Errors**:
- Raises `ModuleError` if divisor is zero

```julia
#| module: Divide
#| input_ports:
#|   - name: arg1
#|     signature: basic:Float
#|     label: dividend
#|   - name: arg2
#|     signature: basic:Float
#|     label: divisor
#| output_ports:
#|   - name: result
#|     signature: basic:Float
#|     label: quotient

# Get inputs
arg1 = get_input("arg1")
arg2 = get_input("arg2")

# Check for division by zero
if arg2 == 0.0
    throw(ModuleError("Division by zero"))
end

# Compute and set output
set_output("result", arg1 / arg2)
```

## Add Module

Adds two floating-point numbers.

**Inputs**:
- `a` (Float): first operand
- `b` (Float): second operand

**Outputs**:
- `sum` (Float): result

```julia
#| module: Add
#| input_ports:
#|   - name: a
#|     signature: basic:Float
#|     label: first operand
#|   - name: b
#|     signature: basic:Float
#|     label: second operand
#| output_ports:
#|   - name: sum
#|     signature: basic:Float
#|     label: result

set_output("sum", get_input("a") + get_input("b"))
```

## Multiply Module

Multiplies two floating-point numbers.

**Inputs**:
- `a` (Float): first factor
- `b` (Float): second factor

**Outputs**:
- `product` (Float): result

```julia
#| module: Multiply
#| input_ports:
#|   - name: a
#|     signature: basic:Float
#|     label: first factor
#|   - name: b
#|     signature: basic:Float
#|     label: second factor
#| output_ports:
#|   - name: product
#|     signature: basic:Float
#|     label: result

set_output("product", get_input("a") * get_input("b"))
```
