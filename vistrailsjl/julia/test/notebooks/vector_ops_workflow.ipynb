# Vector Operations Workflow

This workflow tests the vector operation modules from the Control Flow package.

Tests:
1. **Sum** - Sum elements in a list: [1, 2, 3, 4] → 10
2. **Cross** - Cross product: [1, 2, -1] × [0, 2, 5] → [12, -5, 2]
3. **Dot** - Dot product: [2, 0, -1] · [4, 2, 3] → 5
4. **ElementwiseProduct** (numerical) - [1, 2, 3] ⊙ [2, 0, -1] → [2, 0, -3]
5. **ElementwiseProduct** (tuple) - [1, 2, 3] ⊙ [2, 0, -1] → [(1,2), (2,0), (3,-1)]
