# HTTP Package

A package with HTTPFile module for fetching data from URLs.

```julia
#| package-meta
#| identifier: org.vistrails.vistrails.http
#| version: 1.0.0
```

## HTTPFile Module

Fetches content from an HTTP/HTTPS URL.

**Parameters**:
- `url::String` - The URL to fetch

**Outputs**:
- `file::String` - The content of the fetched file

```julia
#| module: HTTPFile
#| output_ports:
#|   - name: file
#|     signature: basic:String
#| parameters:
#|   - name: url
#|     signature: basic:String

using HTTP

url = get_parameter("url")
println("Fetching: ", url)

try
    response = HTTP.get(url)
    content = String(response.body)
    set_output("file", content)
    println("  ✓ Fetched $(length(content)) bytes")
catch e
    println("  ✗ Error: ", e)
    rethrow(e)
end
```

## JSONParser Module

Parses JSON content and extracts a field.

**Inputs**:
- `json_string::String` - Raw JSON text

**Parameters**:
- `field::String` - Field to extract from JSON

**Outputs**:
- `result::String` - Extracted field value

```julia
#| module: JSONParser
#| input_ports:
#|   - name: json_string
#|     signature: basic:String
#| output_ports:
#|   - name: result
#|     signature: basic:String
#| parameters:
#|   - name: field
#|     signature: basic:String

using JSON

json_str = get_input("json_string")
field = get_parameter("field")

println("Parsing JSON and extracting field: ", field)

data = JSON.parse(json_str)
value = string(data[field])
set_output("result", value)

println("  ✓ Extracted: ", value)
```
