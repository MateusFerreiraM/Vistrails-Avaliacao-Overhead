# Branching Test Package

This package provides utility modules for testing branching workflows:
- **StringSource**: Generates a string
- **StringLength**: Computes string length
- **StringUpper**: Converts to uppercase
- **StringReverse**: Reverses a string
- **CombineStrings**: Concatenates two strings

## Package Metadata

```julia
#| package-meta
#| identifier: org.vistrails.vistrails.branching
#| version: 1.0.0
#| name: Branching Test Package
#| description: Utility modules for testing workflow branching
```

## StringSource Module

Generates a string output. Used as the source node in branching tests.

```julia
#| module: StringSource
#| output_ports:
#|   - name: value
#|     signature: basic:String
#|     label: output string
#| parameters:
#|   - name: text
#|     signature: basic:String

text = get_parameter("text")
set_output("value", text)
```

## StringLength Module

Computes the length of a string.

```julia
#| module: StringLength
#| input_ports:
#|   - name: input
#|     signature: basic:String
#|     label: input string
#| output_ports:
#|   - name: length
#|     signature: basic:Integer
#|     label: string length

input_str = get_input("input")
set_output("length", length(input_str))
```

## StringUpper Module

Converts a string to uppercase.

```julia
#| module: StringUpper
#| input_ports:
#|   - name: input
#|     signature: basic:String
#|     label: input string
#| output_ports:
#|   - name: result
#|     signature: basic:String
#|     label: uppercase string

input_str = get_input("input")
set_output("result", uppercase(input_str))
```

## StringReverse Module

Reverses a string.

```julia
#| module: StringReverse
#| input_ports:
#|   - name: input
#|     signature: basic:String
#|     label: input string
#| output_ports:
#|   - name: result
#|     signature: basic:String
#|     label: reversed string

input_str = get_input("input")
set_output("result", reverse(input_str))
```

## CombineStrings Module

Concatenates two strings with a separator.

```julia
#| module: CombineStrings
#| input_ports:
#|   - name: str1
#|     signature: basic:String
#|     label: first string
#|   - name: str2
#|     signature: basic:String
#|     label: second string
#| output_ports:
#|   - name: result
#|     signature: basic:String
#|     label: combined string
#| parameters:
#|   - name: separator
#|     signature: basic:String

str1 = get_input("str1")
str2 = get_input("str2")
sep = get_parameter("separator")
set_output("result", str1 * sep * str2)
```
