# Optional Inputs Workflow Test

This workflow demonstrates modules with optional inputs.

## Workflow Structure

```
source1 (10.0) ──→ add1 (no b input) ──→ 10.0

source2 (5.0) ──→ add2 ──→ 12.0 (5 + 7)
source3 (7.0) ──┘

source4 (20.0) ──→ mult1 (no factor) ──→ 20.0

source5 (3.0) ──→ mult2 ──→ 15.0 (3 * 5)
source6 (5.0) ──┘

source7 (42.0) ──→ format1 (no prefix/suffix) ──→ "42.0"

source8 (99.0) ────────┐
prefix_src ("$") ───→ format2 ──→ "$99.0"

source9 (88.0) ────────┐
suffix_src (" km") ──→ format3 ──→ "88.0 km"

source10 (77.0) ───────┐
prefix2 ("[") ────────→ format4 ──→ "[77.0]"
suffix2 ("]") ────────┘
```

Tests:
1. OptionalAdd with no second input (defaults to 0)
2. OptionalAdd with both inputs
3. OptionalMultiply with no factor (defaults to 1)
4. OptionalMultiply with factor
5. FormatNumber with no prefix/suffix
6. FormatNumber with prefix only
7. FormatNumber with suffix only
8. FormatNumber with both prefix and suffix
