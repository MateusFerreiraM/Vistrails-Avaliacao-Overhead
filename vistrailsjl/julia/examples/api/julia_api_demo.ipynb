# VisTrailsJL High-Level API Demo

This notebook demonstrates the high-level API for VisTrailsJL, which is similar to the Python VisTrails API.

The API provides a user-friendly way to:
- Load and execute .vt files
- Load and execute notebook workflows
- Navigate version trees
- Parameterize workflows
- Extract outputs

```julia
# Load the VisTrailsJL module
include("../../src/VisTrailsJL.jl")
using .VisTrailsJL
```

## Loading .vt Files

The `load_vistrail()` function loads a .vt file and returns a `VistrailWrapper` object that provides convenient methods for working with workflows.

```julia
# Load a vistrail file
vt = load_vistrail("../../examples/gcd.vt")
println(vt)
```

```julia
# The latest version is selected by default
println("Current version: ", vt.vistrail.current_version)
println("Total versions: ", length(vt.vistrail.actions))
```

## Executing Workflows

Use the `execute()` function to run a workflow and get results.

```julia
# Execute the workflow
result = execute(vt)
println(result)
```

## Extracting Outputs

The `ExecutionResult` object provides methods to access module outputs.

```julia
# Show all executed modules
for (mod_id, outputs) in result.cache
    mod = get_module(result.pipeline, mod_id)
    println("Module $mod_id ($(mod.descriptor.name)):")
    for (key, value) in outputs
        println("  $key = $value")
    end
end
```

```julia
# Get output from a specific module by ID
# (Replace 1 with an actual module ID from your workflow)
if !isempty(result.cache)
    first_mod_id = first(keys(result.cache))
    outputs = result.cache[first_mod_id]
    if !isempty(outputs)
        first_port = first(keys(outputs))
        value = module_output(result, first_mod_id, first_port)
        println("Module $first_mod_id, port '$first_port': ", value)
    end
end
```

You can query information about module types using `describe_module()`.

```julia
# Describe a module type
describe_module("basic:String")
```

```julia
# Describe another module
describe_module("julia:JuliaSource")
```

## Summary

The VisTrailsJL high-level API provides:

### Loading
- `load_vistrail(path)` - Load .vt files
- `load_workflow(path)` - Load notebook workflows  
- `load_package(path)` - Load package notebooks

### Execution
- `execute(vt/workflow)` - Execute workflows
- Parameter injection via keyword arguments (coming soon)

### Output Access
- `output_port(result, name)` - Get OutputPort values
- `module_output(result, id/name, port)` - Get any module's outputs

### Introspection
- `describe_module(type)` - Show module capabilities
- `get_module(pipeline, name)` - Find modules

This API is designed to be compatible with the Python VisTrails API style, making it familiar to existing VisTrails users!
