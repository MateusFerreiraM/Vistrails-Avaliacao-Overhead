# Example: HTTP Data Fetching Workflow

This notebook demonstrates fetching data from an HTTP endpoint and processing it.

**Workflow Steps:**
1. Fetch JSON data from a public API
2. Parse and extract information
3. Display results

**What you'll learn:**
- How to use HTTPFile module to fetch web data
- How to process JSON data in JuliaSource
- How to chain modules together
- Real-world workflow integration

## Define the Workflow

````julia
#| workflow: http_fetch
#| version: 1
````

## Step 1: Fetch Data from API

We'll use a public API that provides random user data for testing.
API: https://randomuser.me/api/

````julia
#| module-id: fetch_user_data
#| module-type: basic:HTTPFile
#| params:
#|   - url: "https://randomuser.me/api/"
````

## Step 2: Parse JSON Response

Extract user information from the JSON response.

````julia
#| module-id: parse_user
#| module-type: julia:JuliaSource
#| inputs:
#|   - json_data: fetch_user_data.file

using JSON

json_str = get_input("json_data")
data = JSON.parse(json_str)

# Extract first user from results
user = data["results"][1]
name = user["name"]
location = user["location"]

user_info = Dict(
    "full_name" => "$(name["title"]) $(name["first"]) $(name["last"])",
    "email" => user["email"],
    "city" => location["city"],
    "country" => location["country"]
)

set_output("self", user_info)
````

## Step 3: Format and Display

Create a nicely formatted display of the user information.

````julia
#| module-id: format_output
#| module-type: julia:JuliaSource
#| inputs:
#|   - user: parse_user.self

user = get_input("user")

output = """
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Random User Data
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
Name:     $(user["full_name"])
Email:    $(user["email"])
Location: $(user["city"]), $(user["country"])
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
"""

set_output("self", output)
````

````julia
#| module-id: display
#| module-type: basic:StandardOutput
#| inputs:
#|   - value: format_output.self
````
