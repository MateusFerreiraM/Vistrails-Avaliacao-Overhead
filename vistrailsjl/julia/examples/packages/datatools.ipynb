## CSVParser

Parse CSV data from a string.

**Inputs**:
- `data::String` - CSV data as string
- `delimiter::String` (optional) - Field delimiter (default: ",")

**Outputs**:
- `rows::Vector` - Parsed rows as vectors
- `num_rows::Integer` - Number of rows

```julia
#| module: CSVParser
#| input_ports:
#|   - name: data
#|     signature: basic:String
#|   - name: delimiter
#|     signature: basic:String
#|     optional: true
#| output_ports:
#|   - name: rows
#|     signature: basic:List
#|   - name: num_rows
#|     signature: basic:Integer

# Get inputs
data = get_input("data")
delimiter = has_input("delimiter") ? get_input("delimiter") : ","

# Parse CSV
lines = split(strip(data), "\n")
rows = [split(line, delimiter) for line in lines]

# Set outputs
set_output("rows", rows)
set_output("num_rows", length(rows))
```

## FilterRows

Filter rows based on a condition.

**Inputs**:
- `rows::Vector` - Input rows
- `column::Integer` - Column index (1-based)
- `min_value::Float` (optional) - Minimum value for filtering

**Outputs**:
- `filtered::Vector` - Filtered rows
- `count::Integer` - Number of filtered rows

```julia
#| module: FilterRows
#| input_ports:
#|   - name: rows
#|     signature: basic:List
#|   - name: column
#|     signature: basic:Integer
#|   - name: min_value
#|     signature: basic:Float
#|     optional: true
#| output_ports:
#|   - name: filtered
#|     signature: basic:List
#|   - name: count
#|     signature: basic:Integer

# Get inputs
rows = get_input("rows")
column = get_input("column")
min_value = has_input("min_value") ? get_input("min_value") : 0.0

# Filter rows
filtered = filter(rows) do row
    if column <= length(row)
        value = tryparse(Float64, row[column])
        value !== nothing && value >= min_value
    else
        false
    end
end

# Set outputs
set_output("filtered", filtered)
set_output("count", length(filtered))
```

## ComputeStats

Compute statistics on numeric data.

**Inputs**:
- `values::Vector` - Numeric values

**Outputs**:
- `mean::Float` - Mean value
- `min::Float` - Minimum value
- `max::Float` - Maximum value

```julia
#| module: ComputeStats
#| input_ports:
#|   - name: values
#|     signature: basic:List
#| output_ports:
#|   - name: mean
#|     signature: basic:Float
#|   - name: min
#|     signature: basic:Float
#|   - name: max
#|     signature: basic:Float

# Get input
values = get_input("values")

# Compute statistics
using Statistics
mean_val = mean(values)
min_val = minimum(values)
max_val = maximum(values)

# Set outputs
set_output("mean", mean_val)
set_output("min", min_val)
set_output("max", max_val)
```
