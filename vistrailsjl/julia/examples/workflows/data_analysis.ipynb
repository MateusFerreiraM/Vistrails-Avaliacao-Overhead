# Data Analysis Workflow

Process CSV data and compute statistics using custom DataTools package.

```julia
#| workflow: data_analysis
#| version: 1
#| description: CSV data processing with filtering and statistics
```

## Step 1: Generate Sample Data

Create sample CSV data with temperature readings.

**Outputs:**
- `data` (String) - CSV formatted data

```julia
#| module-id: generate_data
#| module-type: julia:JuliaSource

# Generate sample CSV data
csv_data = """Day,Temperature
Monday,23.5
Tuesday,18.2
Wednesday,25.8
Thursday,22.1
Friday,19.7
Saturday,24.3
Sunday,21.5"""

set_output("data", csv_data)
```

## Step 2: Parse CSV

Use the DataTools CSVParser to parse the data.

**Outputs:**
- `rows` (List) - Parsed CSV rows as vectors
- `num_rows` (Integer) - Number of rows parsed

```julia
#| module-id: parse_csv
#| module-type: datatools:CSVParser
#| inputs:
#|   data: generate_data.data
```

## Step 3: Extract Temperature Column

Extract temperature values from parsed rows.

**Outputs:**
- `temperatures` (List) - Temperature values as floats

```julia
#| module-id: extract_temps
#| module-type: julia:JuliaSource
#| inputs:
#|   rows: parse_csv.rows

# Get rows and skip header
rows = get_input("rows")
data_rows = rows[2:end]  # Skip header

# Extract temperature column (index 2)
temps = [parse(Float64, row[2]) for row in data_rows]

set_output("temperatures", temps)
```

## Step 4: Compute Statistics

Use DataTools ComputeStats to calculate mean, min, max.

**Outputs:**
- `mean` (Float) - Mean temperature
- `min` (Float) - Minimum temperature  
- `max` (Float) - Maximum temperature

```julia
#| module-id: compute_stats
#| module-type: datatools:ComputeStats
#| inputs:
#|   values: extract_temps.temperatures
```

## Step 5: Display Results

Print the computed statistics.

**Outputs:**
- `summary` (String) - Summary text with all statistics

```julia
#| module-id: display_results
#| module-type: julia:JuliaSource
#| inputs:
#|   mean: compute_stats.mean
#|   min: compute_stats.min
#|   max: compute_stats.max

mean_temp = get_input("mean")
min_temp = get_input("min")
max_temp = get_input("max")

println("Temperature Statistics:")
println("  Mean: $mean_temp$(mean_temp)$(min_temp)$(max_temp)°C")
println("  Min:  $min_temp$(mean_temp)$(min_temp)$(max_temp)°C")
println("  Max:  $max_temp$(mean_temp)$(min_temp)$(max_temp)°C")

set_output("summary", "Mean: $mean_temp, Min: $min_temp, Max: $max_temp")
```
