# Matplotlib Line Plot Example

This notebook demonstrates a simple matplotlib-style line plot workflow in VisTrailsJL.

Equivalent to examples/matplotlib/lineplot_ex3.vt from Python VisTrails.

```julia
#| workflow: lineplot_example
#| version: 1
#| description: Simple line plot with matplotlib modules
```

## Define the Line Plot

MplLinePlot module with x and y data

```julia
#| module-id: lineplot
#| module-type: matplotlib:MplLinePlot
#| params:
#|   x: "[1, 2, 3, 4]"
#|   y: "[20, 21, 20.5, 20.8]"
```

## Create Figure Container

MplFigure receives plot functions and builds the composite figure

```julia
#| module-id: figure
#| module-type: matplotlib:MplFigure
#| inputs:
#|   addPlot: lineplot.value
```

## Output the Figure

MplFigureOutput saves the figure to a PNG file

```julia
#| module-id: output
#| module-type: matplotlib:MplFigureOutput
#| inputs:
#|   value: figure.self
```

## Execute Workflow

This directive tells VisTrailsJL to execute the workflow defined above

```julia
#| execute
```

## Workflow Summary

**Modules:** 3
- MplLinePlot (id: lineplot) - defines the line plot with x and y data
- MplFigure (id: figure) - container that builds the plot
- MplFigureOutput (id: output) - saves to matplotlib_output.png

**Connections:** 2
- lineplot.value → figure.addPlot
- figure.self → output.value

**Output:** matplotlib_output.png (line plot with 4 points)

**Data Flow:**
1. MplLinePlot creates a function that will plot x=[1,2,3,4], y=[20,21,20.5,20.8]
2. MplFigure receives this function, creates a Plots.jl figure, calls the function
3. MplFigureOutput receives the figure and saves it to PNG

This demonstrates the key VisTrails pattern:
- Plot modules output **functions** (lambdas), not data
- MplFigure **calls** these functions to build the composite plot
- This allows multiple plots to be combined on one figure
