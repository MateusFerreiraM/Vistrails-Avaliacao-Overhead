# Example: Data Analysis Workflow

This notebook demonstrates a simple data analysis workflow using VisTrailsJL.

**Workflow Steps:**
1. Generate sample data (list of numbers)
2. Calculate statistics (sum, mean, max)
3. Display results

**What you'll learn:**
- How to define workflows in notebooks
- How to use JuliaSource modules for custom computation
- How to use List and StandardOutput modules
- How connections pass data between modules

## Define the Workflow

````julia
#| workflow: data_analysis
#| version: 1
````

## Step 1: Generate Sample Data

We'll create a list of 10 random numbers using a JuliaSource module.

````julia
#| module-id: generate_data
#| module-type: julia:JuliaSource

# Generate 10 random numbers between 1 and 100
data = [rand(1:100) for _ in 1:10]
set_output("self", data)
````

## Step 2: Calculate Sum

Use the built-in Sum module to add all numbers.

````julia
#| module-id: calc_sum
#| module-type: control_flow:Sum
#| inputs:
#|   - InputList: generate_data.self
````

## Step 3: Calculate Mean

Custom JuliaSource module to calculate the mean (average).

````julia
#| module-id: calc_mean
#| module-type: julia:JuliaSource
#| inputs:
#|   - data: generate_data.self

data = get_input("data")
mean_value = sum(data) / length(data)
set_output("self", mean_value)
````

## Step 4: Find Maximum

Another custom module to find the maximum value.

````julia
#| module-id: calc_max
#| module-type: julia:JuliaSource
#| inputs:
#|   - data: generate_data.self

data = get_input("data")
max_value = maximum(data)
set_output("self", max_value)
````

## Step 5: Display Results

Create a summary report and display it.

````julia
#| module-id: create_report
#| module-type: julia:JuliaSource
#| inputs:
#|   - data: generate_data.self
#|   - total: calc_sum.Result
#|   - mean: calc_mean.self
#|   - max: calc_max.self

data = get_input("data")
total = get_input("total")
mean = get_input("mean")
max_val = get_input("max")

report = """
Data Analysis Report
====================
Data: $data
Count: $(length(data))
Sum: $total
Mean: $(round(mean, digits=2))
Max: $max_val
"""

set_output("self", report)
````

````julia
#| module-id: display
#| module-type: basic:StandardOutput
#| inputs:
#|   - value: create_report.self
````

## Execute the Workflow

The `#| execute` directive tells VisTrailsJL to run this workflow when the notebook is loaded.

````julia
#| execute
````

## How to Run This Workflow

```julia
using VisTrailsJL

# Parse and execute the workflow
workflow = parse_workflow_notebook("data_analysis_workflow.ipynb")
pipeline, id_to_module = build_pipeline_from_workflow(workflow)
cache, outputs = execute_notebook_pipeline(pipeline, workflow; id_to_module=id_to_module)
```
